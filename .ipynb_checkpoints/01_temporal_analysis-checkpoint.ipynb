# Temporal Analysis for Network Traffic

## Time-based Pattern Detection and Traffic Forecasting

**SDN Applications:**
- Predict traffic spikes before they occur
- Enable proactive rerouting and load balancing
- Auto-scale network resources based on predictions

**This notebook includes:**
1. Time window analysis
2. Traffic forecasting (10 steps ahead)
3. Pattern detection
4. Flow sequence analysis

---
## Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
import os
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

print("✓ Libraries imported")
```

```python
# Configuration
DATA_PATH = "network_traffic.csv"  # ← UPDATE THIS
OUTPUT_DIR = "results"
WINDOW_SIZE = 10
FORECAST_STEPS = 10

os.makedirs(OUTPUT_DIR, exist_ok=True)

EXPECTED_FEATURES = ['SourceIP', 'DestinationIP', 'SourcePort', 'DestinationPort',
                    'Protocol', 'BytesSent', 'BytesReceived', 'PacketsSent', 
                    'PacketsReceived', 'Duration']

print(f"✓ Configuration set")
print(f"  Window size: {WINDOW_SIZE}")
print(f"  Forecast steps: {FORECAST_STEPS}")
```

---
## Load Data

```python
# Load dataset
df = pd.read_csv(DATA_PATH)

# Sample if too large
if len(df) > 50000:
    print(f"Sampling {len(df):,} rows to 50,000")
    if 'IsAnomaly' in df.columns:
        sample_df, _ = train_test_split(df, train_size=50000, stratify=df['IsAnomaly'], random_state=42)
        df = sample_df
    else:
        df = df.sample(n=50000, random_state=42)

feature_cols = [col for col in EXPECTED_FEATURES if col in df.columns]

print(f"✓ Data loaded: {df.shape}")
print(f"  Features: {len(feature_cols)}")
df.head()
```

---
## 1. Time Window Analysis

Create sliding windows to detect temporal patterns.

```python
print("Creating time windows...")

# Add time index
df['TimeIndex'] = range(len(df))

# Calculate rolling statistics
temporal_features = pd.DataFrame()
temporal_features['TimeIndex'] = df['TimeIndex']

for col in feature_cols[:5]:  # First 5 features
    temporal_features[f'{col}_mean'] = df[col].rolling(window=WINDOW_SIZE, min_periods=1).mean()
    temporal_features[f'{col}_std'] = df[col].rolling(window=WINDOW_SIZE, min_periods=1).std()
    temporal_features[f'{col}_trend'] = df[col].diff()

# Detect spikes
for col in feature_cols[:3]:
    threshold = df[col].std() * 2
    temporal_features[f'{col}_spike'] = (abs(df[col].diff()) > threshold).astype(int)

print(f"✓ Created {len(temporal_features.columns)} temporal features")
temporal_features.head()
```

```python
# Visualize temporal patterns
fig = plt.figure(figsize=(20, 12))

# Subsample for plotting
max_plot = min(10000, len(temporal_features))
if len(temporal_features) > max_plot:
    plot_idx = np.random.choice(len(temporal_features), max_plot, replace=False)
    plot_idx.sort()
    temp_plot = temporal_features.iloc[plot_idx]
else:
    temp_plot = temporal_features

# 1. Rolling means
ax1 = plt.subplot(3, 2, 1)
for col in feature_cols[:3]:
    if f'{col}_mean' in temp_plot.columns:
        ax1.plot(temp_plot['TimeIndex'], temp_plot[f'{col}_mean'], label=col, alpha=0.7, linewidth=1.5)
ax1.set_title('Temporal Trends (Rolling Mean)', fontsize=14, fontweight='bold')
ax1.set_xlabel('Time Index')
ax1.set_ylabel('Value')
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. Volatility
ax2 = plt.subplot(3, 2, 2)
for col in feature_cols[:3]:
    if f'{col}_std' in temp_plot.columns:
        ax2.plot(temp_plot['TimeIndex'], temp_plot[f'{col}_std'], label=col, alpha=0.7, linewidth=1.5)
ax2.set_title('Volatility Over Time (Rolling Std)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Time Index')
ax2.set_ylabel('Standard Deviation')
ax2.legend()
ax2.grid(True, alpha=0.3)

# 3. Spike detection
ax3 = plt.subplot(3, 2, 3)
spike_cols = [col for col in temp_plot.columns if '_spike' in col]
if spike_cols:
    spike_data = temp_plot[spike_cols].sum(axis=1)
    ax3.plot(temp_plot['TimeIndex'], spike_data, color='red', alpha=0.6, linewidth=1)
    ax3.fill_between(temp_plot['TimeIndex'], spike_data, alpha=0.3, color='red')
ax3.set_title('Detected Traffic Spikes', fontsize=14, fontweight='bold')
ax3.set_xlabel('Time Index')
ax3.set_ylabel('Number of Spikes')
ax3.grid(True, alpha=0.3)

# 4. Trends
ax4 = plt.subplot(3, 2, 4)
for col in feature_cols[:2]:
    if f'{col}_trend' in temp_plot.columns:
        ax4.plot(temp_plot['TimeIndex'], temp_plot[f'{col}_trend'], label=col, alpha=0.7, linewidth=1)
ax4.axhline(y=0, color='black', linestyle='--', alpha=0.5)
ax4.set_title('Rate of Change (Trends)', fontsize=14, fontweight='bold')
ax4.set_xlabel('Time Index')
ax4.set_ylabel('Change')
ax4.legend()
ax4.grid(True, alpha=0.3)

# 5. Summary text
ax5 = plt.subplot(3, 2, 5)
summary = f"""TEMPORAL ANALYSIS

Time Points: {len(temporal_features):,}
Window Size: {WINDOW_SIZE}
Features Analyzed: {len(feature_cols)}

SDN Applications:
• Predict traffic patterns
• Detect unusual sequences
• Optimize flow management
• Enable predictive routing"""
ax5.text(0.5, 0.5, summary, ha='center', va='center', fontsize=12, family='monospace',
        bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.5))
ax5.axis('off')
ax5.set_title('Analysis Summary', fontsize=14, fontweight='bold')

# 6. Spike summary
ax6 = plt.subplot(3, 2, 6)
total_spikes = temporal_features[[col for col in temporal_features.columns if '_spike' in col]].sum().sum()
spike_summary = f"""SPIKE DETECTION

Total Spikes: {int(total_spikes)}
Spike Rate: {total_spikes/len(temporal_features)*100:.2f}%

Threshold: 2 × std deviation

Enables:
• Early congestion detection
• Proactive resource allocation
• Attack pattern recognition"""
ax6.text(0.5, 0.5, spike_summary, ha='center', va='center', fontsize=12, family='monospace',
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.5))
ax6.axis('off')
ax6.set_title('Spike Detection', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/temporal_patterns.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Saved: temporal_patterns.png")
```

---
## 2. Traffic Forecasting

Predict future network traffic using lag features.

```python
print("Training traffic forecasting model...")

# Use first feature for demonstration
target_feature = feature_cols[0]

# Create lag features
lag_features = pd.DataFrame()
for i in range(1, 6):  # 5 lag features
    lag_features[f'lag_{i}'] = df[target_feature].shift(i)

lag_features = lag_features.fillna(0)

# Split data
train_size = int(len(lag_features) * 0.8)
X_train_ts = lag_features[:train_size]
y_train_ts = df[target_feature][:train_size]
X_test_ts = lag_features[train_size:]
y_test_ts = df[target_feature][train_size:]

# Train forecaster
forecaster = RandomForestRegressor(n_estimators=50, random_state=42, n_jobs=-1)
forecaster.fit(X_train_ts, y_train_ts)

# Predictions
predictions = forecaster.predict(X_test_ts)

# Metrics
mse = mean_squared_error(y_test_ts, predictions)
mae = mean_absolute_error(y_test_ts, predictions)

print(f"\n✓ Forecasting Performance:")
print(f"  MSE: {mse:.4f}")
print(f"  MAE: {mae:.4f}")
```

```python
# Future predictions (extrapolation)
last_values = df[target_feature].tail(5).values
future_predictions = []

for _ in range(FORECAST_STEPS):
    next_pred = forecaster.predict([last_values])[0]
    future_predictions.append(next_pred)
    last_values = np.append(last_values[1:], next_pred)

print(f"\n✓ Generated {len(future_predictions)} future predictions")
```

```python
# Visualize forecasting results
fig = plt.figure(figsize=(20, 10))

# 1. Historical predictions vs actual
ax1 = plt.subplot(2, 2, 1)
test_indices = range(len(y_test_ts))
# Subsample for visibility
sample_n = min(1000, len(test_indices))
sample_idx = np.linspace(0, len(test_indices)-1, sample_n, dtype=int)

ax1.plot(sample_idx, y_test_ts.values[sample_idx], label='Actual', linewidth=2, alpha=0.7, color='blue')
ax1.plot(sample_idx, predictions[sample_idx], label='Predicted', linewidth=2, alpha=0.7, color='red', linestyle='--')
ax1.set_title('Traffic Prediction Performance', fontsize=14, fontweight='bold')
ax1.set_xlabel('Time Index')
ax1.set_ylabel(target_feature)
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. Prediction errors
ax2 = plt.subplot(2, 2, 2)
errors = y_test_ts.values - predictions
ax2.hist(errors, bins=50, edgecolor='black', alpha=0.7, color='orange')
ax2.axvline(x=0, color='red', linestyle='--', linewidth=2)
ax2.set_title('Prediction Error Distribution', fontsize=14, fontweight='bold')
ax2.set_xlabel('Prediction Error')
ax2.set_ylabel('Frequency')
ax2.grid(True, alpha=0.3)

# 3. Future predictions
ax3 = plt.subplot(2, 2, 3)
last_n = min(50, len(y_test_ts))
historical_indices = range(-last_n, 0)
future_indices = range(0, len(future_predictions))

ax3.plot(historical_indices, y_test_ts.values[-last_n:], label='Historical', linewidth=2, color='blue')
ax3.plot(future_indices, future_predictions, label='Future Forecast', linewidth=2, color='green', 
        linestyle='--', marker='o')
ax3.axvline(x=0, color='red', linestyle='--', alpha=0.5, label='Present')
ax3.set_title(f'Future Traffic Forecast ({FORECAST_STEPS} steps)', fontsize=14, fontweight='bold')
ax3.set_xlabel('Time Steps (negative=past, positive=future)')
ax3.set_ylabel(target_feature)
ax3.legend()
ax3.grid(True, alpha=0.3)

# 4. SDN Applications
ax4 = plt.subplot(2, 2, 4)
sdn_text = f"""PREDICTIVE SDN APPLICATIONS

✓ Proactive Resource Allocation
  • Scale resources before demand spikes
  • Pre-configure flow tables

✓ Traffic Engineering
  • Route optimization based on predictions
  • Load balancing ahead of congestion

✓ Anomaly Prevention
  • Detect anomalies before they occur
  • Automatic mitigation actions

Performance:
  MSE: {mse:.4f}
  MAE: {mae:.4f}"""

ax4.text(0.5, 0.5, sdn_text, ha='center', va='center', fontsize=11, family='monospace',
        bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.5))
ax4.axis('off')
ax4.set_title('Future SDN Benefits', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/traffic_forecasting.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Saved: traffic_forecasting.png")
```

---
## 3. Export Results

```python
# Export predictions
forecast_results = pd.DataFrame({
    'Actual': y_test_ts.values,
    'Predicted': predictions,
    'Error': y_test_ts.values - predictions
})

forecast_results.to_csv(f'{OUTPUT_DIR}/temporal_predictions.csv', index=False)
print("✓ Saved: temporal_predictions.csv")

# Export future predictions
future_df = pd.DataFrame({
    'Step': range(1, len(future_predictions) + 1),
    'Predicted_Value': future_predictions
})

future_df.to_csv(f'{OUTPUT_DIR}/future_forecast.csv', index=False)
print("✓ Saved: future_forecast.csv")
```
