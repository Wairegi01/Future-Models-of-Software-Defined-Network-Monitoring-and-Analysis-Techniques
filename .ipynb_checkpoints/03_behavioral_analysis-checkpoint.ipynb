# Behavioral Analysis for Network Traffic

## Traffic Profiling and Intelligent Classification

**SDN Applications:**
- Automatic traffic classification into behavior profiles
- Intent-based networking policies
- Anomaly severity scoring for prioritization
- Resource optimization based on profiles

**This notebook includes:**
1. Behavioral profiling (K-Means clustering)
2. Density-based outlier detection (DBSCAN)
3. Network flow analysis
4. Anomaly severity scoring

---
## Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, precision_score, recall_score
from collections import Counter
import os
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

print("✓ Libraries imported")
```

```python
# Configuration
DATA_PATH = "network_traffic.csv"  # ← UPDATE THIS
OUTPUT_DIR = "results"
N_PROFILES = 5  # Number of behavior profiles
DBSCAN_EPS = 0.5
DBSCAN_MIN_SAMPLES = 5

os.makedirs(OUTPUT_DIR, exist_ok=True)

EXPECTED_FEATURES = ['SourceIP', 'DestinationIP', 'SourcePort', 'DestinationPort',
                    'Protocol', 'BytesSent', 'BytesReceived', 'PacketsSent', 
                    'PacketsReceived', 'Duration']

print(f"✓ Configuration set")
print(f"  Number of profiles: {N_PROFILES}")
print(f"  DBSCAN eps: {DBSCAN_EPS}")
```

---
## Load Data

```python
# Load dataset
df = pd.read_csv(DATA_PATH)

# Sample if too large
if len(df) > 50000:
    print(f"Sampling {len(df):,} rows to 50,000")
    if 'IsAnomaly' in df.columns:
        sample_df, _ = train_test_split(df, train_size=50000, stratify=df['IsAnomaly'], random_state=42)
        df = sample_df
    else:
        df = df.sample(n=50000, random_state=42)

feature_cols = [col for col in EXPECTED_FEATURES if col in df.columns]
X = df[feature_cols].values

print(f"✓ Data loaded: {df.shape}")
print(f"  Features: {len(feature_cols)}")
```

---
## 1. Behavioral Profiling

Cluster traffic into natural behavior profiles using K-Means.

```python
print(f"Creating {N_PROFILES} behavioral profiles...")

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# K-Means clustering
kmeans = KMeans(n_clusters=N_PROFILES, random_state=42, n_init=10)
profiles = kmeans.fit_predict(X_scaled)

# Analyze each profile
print(f"\n✓ Created {N_PROFILES} profiles")
print("\nProfile Summary:")

for i in range(N_PROFILES):
    mask = profiles == i
    count = mask.sum()
    pct = count / len(profiles) * 100
    
    profile_data = df[mask][feature_cols]
    means = profile_data.mean()
    
    print(f"\nProfile {i}: {count:,} samples ({pct:.1f}%)")
    print(f"  Top characteristics:")
    top_features = means.abs().sort_values(ascending=False).head(3)
    for feat, val in top_features.items():
        print(f"    {feat}: {val:.3f}")
    
    if 'IsAnomaly' in df.columns:
        anomaly_rate = df[mask]['IsAnomaly'].mean()
        print(f"  Anomaly rate: {anomaly_rate*100:.1f}%")
```

```python
# Visualize behavioral profiles
fig = plt.figure(figsize=(20, 12))

# 1. Profile distribution
ax1 = plt.subplot(2, 3, 1)
profile_counts = Counter(profiles)
colors = plt.cm.Set3(range(N_PROFILES))
ax1.bar(range(N_PROFILES), [profile_counts[i] for i in range(N_PROFILES)], 
       color=colors, edgecolor='black')
ax1.set_title('Behavioral Profile Distribution', fontsize=14, fontweight='bold')
ax1.set_xlabel('Profile ID')
ax1.set_ylabel('Count')
ax1.grid(True, alpha=0.3, axis='y')

# 2. PCA visualization of profiles
ax2 = plt.subplot(2, 3, 2)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Subsample for plotting
max_plot = min(10000, len(X_pca))
plot_idx = np.random.choice(len(X_pca), max_plot, replace=False) if len(X_pca) > max_plot else np.arange(len(X_pca))

for i in range(N_PROFILES):
    mask = profiles[plot_idx] == i
    ax2.scatter(X_pca[plot_idx][mask, 0], X_pca[plot_idx][mask, 1], 
               label=f'Profile {i}', alpha=0.6, s=10, color=colors[i])
ax2.set_title('Profiles in PCA Space', fontsize=14, fontweight='bold')
ax2.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%})')
ax2.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%})')
ax2.legend()

# 3. Profile characteristics heatmap
ax3 = plt.subplot(2, 3, 3)
char_matrix = []
for i in range(N_PROFILES):
    mask = profiles == i
    means = df[mask][feature_cols].mean().values[:8]  # First 8 features
    char_matrix.append(means)
char_matrix = np.array(char_matrix)

sns.heatmap(char_matrix, annot=False, cmap='coolwarm', 
           xticklabels=feature_cols[:8],
           yticklabels=[f'P{i}' for i in range(N_PROFILES)],
           ax=ax3, cbar_kws={'label': 'Value'})
ax3.set_title('Profile Characteristics', fontsize=14, fontweight='bold')

# 4. Anomaly rate by profile
if 'IsAnomaly' in df.columns:
    ax4 = plt.subplot(2, 3, 4)
    anomaly_rates = []
    for i in range(N_PROFILES):
        mask = profiles == i
        rate = df[mask]['IsAnomaly'].mean()
        anomaly_rates.append(rate)
    
    ax4.bar(range(N_PROFILES), anomaly_rates, color=colors, edgecolor='black')
    ax4.set_title('Anomaly Rate by Profile', fontsize=14, fontweight='bold')
    ax4.set_xlabel('Profile ID')
    ax4.set_ylabel('Anomaly Rate')
    ax4.grid(True, alpha=0.3, axis='y')
    
    for i, rate in enumerate(anomaly_rates):
        ax4.text(i, rate + 0.01, f'{rate*100:.1f}%', ha='center', fontweight='bold')

# 5. SDN Applications
ax5 = plt.subplot(2, 3, 5)
sdn_text = """BEHAVIORAL PROFILING SDN USES

✓ Intent-Based Networking
  • Auto-classify traffic types
  • Generate QoS policies

✓ Baseline Learning
  • Normal behavior per profile
  • Detect deviations

✓ Traffic Engineering
  • Route optimization per profile
  • Resource allocation

✓ Security Policies
  • Profile-specific rules
  • Anomaly thresholds"""
ax5.text(0.5, 0.5, sdn_text, ha='center', va='center', fontsize=11, family='monospace',
        bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.5))
ax5.axis('off')
ax5.set_title('SDN Applications', fontsize=14, fontweight='bold')

# 6. Summary
ax6 = plt.subplot(2, 3, 6)
summary = f"""PROFILING SUMMARY

Total Samples: {len(profiles):,}
Behavior Profiles: {N_PROFILES}

Largest Profile: {max(profile_counts.values()):,}
Smallest Profile: {min(profile_counts.values()):,}
"""
ax6.text(0.5, 0.5, summary, ha='center', va='center', fontsize=12, family='monospace',
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.5))
ax6.axis('off')
ax6.set_title('Summary', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/behavioral_profiling.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Saved: behavioral_profiling.png")
```

---
## 2. Density-Based Clustering (DBSCAN)

Find outliers that don't fit any cluster.

```python
print(f"Running DBSCAN (eps={DBSCAN_EPS}, min_samples={DBSCAN_MIN_SAMPLES})...")

# DBSCAN clustering
dbscan = DBSCAN(eps=DBSCAN_EPS, min_samples=DBSCAN_MIN_SAMPLES, n_jobs=-1)
clusters = dbscan.fit_predict(X_scaled)

# Analyze results
n_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
n_outliers = (clusters == -1).sum()
outlier_pct = n_outliers / len(clusters) * 100

print(f"\n✓ Clustering complete")
print(f"  Number of clusters: {n_clusters}")
print(f"  Outliers detected: {n_outliers:,} ({outlier_pct:.2f}%)")

cluster_counts = Counter(clusters)
print(f"\n  Cluster sizes:")
for cluster_id, count in sorted(cluster_counts.items()):
    if cluster_id == -1:
        print(f"    Outliers: {count:,}")
    else:
        print(f"    Cluster {cluster_id}: {count:,}")

if 'IsAnomaly' in df.columns:
    outlier_labels = (clusters == -1).astype(int)
    acc = accuracy_score(df['IsAnomaly'], outlier_labels)
    prec = precision_score(df['IsAnomaly'], outlier_labels, zero_division=0)
    rec = recall_score(df['IsAnomaly'], outlier_labels)
    
    print(f"\n  Performance vs true anomalies:")
    print(f"    Accuracy:  {acc:.4f}")
    print(f"    Precision: {prec:.4f}")
    print(f"    Recall:    {rec:.4f}")
```

```python
# Visualize DBSCAN results
fig = plt.figure(figsize=(18, 10))

# 1. Clusters in PCA space
ax1 = plt.subplot(2, 3, 1)
unique_clusters = sorted(set(clusters))
colors_db = plt.cm.Spectral(np.linspace(0, 1, len(unique_clusters)))

# Subsample for plotting
max_plot = min(10000, len(X_pca))
plot_idx = np.random.choice(len(X_pca), max_plot, replace=False) if len(X_pca) > max_plot else np.arange(len(X_pca))

for i, cluster_id in enumerate(unique_clusters):
    mask = clusters[plot_idx] == cluster_id
    if cluster_id == -1:
        ax1.scatter(X_pca[plot_idx][mask, 0], X_pca[plot_idx][mask, 1], 
                   c='red', s=20, alpha=0.8, label='Outliers', marker='x')
    else:
        ax1.scatter(X_pca[plot_idx][mask, 0], X_pca[plot_idx][mask, 1], 
                   c=[colors_db[i]], s=10, alpha=0.6, label=f'Cluster {cluster_id}')

ax1.set_title('DBSCAN Clustering Results', fontsize=14, fontweight='bold')
ax1.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%})')
ax1.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%})')
ax1.legend()

# 2. Cluster sizes
ax2 = plt.subplot(2, 3, 2)
cluster_ids = sorted([c for c in cluster_counts.keys() if c != -1])
counts_c = [cluster_counts[c] for c in cluster_ids]

ax2.bar(range(len(cluster_ids)), counts_c, color='steelblue', edgecolor='black')
if -1 in cluster_counts:
    ax2.bar([len(cluster_ids)], [cluster_counts[-1]], 
           color='red', edgecolor='black', label='Outliers')
ax2.set_xticks(range(len(cluster_ids) + (1 if -1 in cluster_counts else 0)))
ax2.set_xticklabels(cluster_ids + ([-1] if -1 in cluster_counts else []))
ax2.set_title('Cluster Size Distribution', fontsize=14, fontweight='bold')
ax2.set_xlabel('Cluster ID (-1 = Outliers)')
ax2.set_ylabel('Count')
ax2.grid(True, alpha=0.3, axis='y')

# 3. Confusion matrix (if labels available)
if 'IsAnomaly' in df.columns:
    ax3 = plt.subplot(2, 3, 3)
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(df['IsAnomaly'], outlier_labels)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges', ax=ax3,
                xticklabels=['Not Outlier', 'Outlier'],
                yticklabels=['Normal', 'Anomaly'])
    ax3.set_title('Outlier Detection Performance', fontsize=14, fontweight='bold')
    ax3.set_ylabel('True Label')
    ax3.set_xlabel('DBSCAN Prediction')

# 4. Outlier percentage
ax4 = plt.subplot(2, 3, 4)
labels_pie = ['Clustered', 'Outliers']
sizes_pie = [len(clusters) - n_outliers, n_outliers]
colors_pie = ['lightblue', 'red']
ax4.pie(sizes_pie, labels=labels_pie, colors=colors_pie, autopct='%1.1f%%', startangle=90)
ax4.set_title('Outlier Proportion', fontsize=14, fontweight='bold')

# 5. SDN Benefits
ax5 = plt.subplot(2, 3, 5)
sdn_text = """DENSITY-BASED SDN USES

✓ Rare Event Detection
  • Port scanning
  • Network probing
  • DDoS precursors

✓ Flexible Boundaries
  • No fixed anomaly threshold
  • Adapts to data shape

✓ Noise Filtering
  • Separate true anomalies
  • From random fluctuations"""
ax5.text(0.5, 0.5, sdn_text, ha='center', va='center', fontsize=11, family='monospace',
        bbox=dict(boxstyle='round', facecolor='lightcoral', alpha=0.5))
ax5.axis('off')
ax5.set_title('SDN Benefits', fontsize=14, fontweight='bold')

# 6. Summary
ax6 = plt.subplot(2, 3, 6)
summary = f"""DBSCAN SUMMARY

Total Samples: {len(clusters):,}
Clusters Found: {n_clusters}
Outliers: {n_outliers:,} ({outlier_pct:.1f}%)

Outliers are potential:
• Attack traffic
• Misconfigurations
• Rare normal events"""
ax6.text(0.5, 0.5, summary, ha='center', va='center', fontsize=12, family='monospace',
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.5))
ax6.axis('off')
ax6.set_title('Summary', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/density_clustering.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Saved: density_clustering.png")
```

---
## 3. Export Results

```python
# Export behavioral profiles
profile_results = pd.DataFrame({
    'Profile': profiles,
    'DBSCAN_Cluster': clusters,
    'Is_Outlier': (clusters == -1).astype(int)
})

if 'IsAnomaly' in df.columns:
    profile_results['True_Label'] = df['IsAnomaly'].values

profile_results.to_csv(f'{OUTPUT_DIR}/behavioral_analysis.csv', index=False)
print("✓ Saved: behavioral_analysis.csv")
```
