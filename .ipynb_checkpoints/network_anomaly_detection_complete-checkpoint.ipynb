## 1. Setup & Configuration

Run this first to import all libraries and set configuration.

```python
# Install required packages (uncomment if needed)
# !pip install pandas numpy matplotlib seaborn scikit-learn --break-system-packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, IsolationForest, GradientBoostingClassifier
from sklearn.metrics import (classification_report, confusion_matrix, 
                           roc_auc_score, roc_curve, precision_recall_curve,
                           f1_score, accuracy_score, precision_score, recall_score,
                           mean_squared_error, mean_absolute_error)
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN, KMeans
from collections import Counter
from datetime import datetime
import os
import warnings
warnings.filterwarnings('ignore')

# Set style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

print("✓ All libraries imported successfully!")
```

```python
# CONFIGURATION - Update these paths!

DATA_PATH = "C:\\Users\\iamwa\\source\\repos\\Thesis\\Network Analysis\\Dataset\\synthetic_network_traffic.csv"  # ← CHANGE THIS to your CSV path
OUTPUT_DIR = "C:\\Users\\iamwa\\source\\repos\\Thesis\\Network Analysis\\Results"
    

# Create output directory
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Model parameters
RF_N_ESTIMATORS = 100
RF_MAX_DEPTH = 10
ISO_CONTAMINATION = 0.1
TEST_SIZE = 0.2
RANDOM_STATE = 42

# Expected features
EXPECTED_FEATURES = ['SourceIP', 'DestinationIP', 'SourcePort', 'DestinationPort',
                    'Protocol', 'BytesSent', 'BytesReceived', 'PacketsSent', 
                    'PacketsReceived', 'Duration']

print(f"✓ Configuration set")
print(f"  Data path: {DATA_PATH}")
print(f"  Output directory: {OUTPUT_DIR}")
```

---
## 2. Data Loading & EDA

Load the dataset and perform exploratory data analysis.

```python
# Load dataset
print("=" * 80)
print("LOADING DATASET")
print("=" * 80)

df = pd.read_csv(DATA_PATH)

# Limit size to prevent memory issues
if len(df) > 50000:
    print(f"\n⚠ Dataset has {len(df):,} rows - sampling to 50,000")
    if 'IsAnomaly' in df.columns:
        # Stratified sampling
        sample_df, _ = train_test_split(
            df, train_size=50000, stratify=df['IsAnomaly'], random_state=RANDOM_STATE
        )
        df = sample_df
    else:
        df = df.sample(n=50000, random_state=RANDOM_STATE)

print(f"\nDataset Shape: {df.shape}")
print(f"\nColumns: {list(df.columns)}")
print(f"\nFirst 5 rows:")
df.head()
```

```python
# Basic statistics
print("Data Types:")
print(df.dtypes)
print("\n" + "="*80)

print("\nMissing Values:")
print(df.isnull().sum())
print("\n" + "="*80)

print("\nBasic Statistics:")
df.describe()
```

```python
# Class distribution
if 'IsAnomaly' in df.columns:
    print("Class Distribution:")
    print(df['IsAnomaly'].value_counts())
    print(f"\nAnomaly Percentage: {df['IsAnomaly'].mean() * 100:.2f}%")
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    df['IsAnomaly'].value_counts().plot(kind='bar', ax=axes[0], color=['#2ecc71', '#e74c3c'])
    axes[0].set_title('Class Distribution', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('IsAnomaly (0=Normal, 1=Anomaly)')
    axes[0].set_ylabel('Count')
    axes[0].set_xticklabels(['Normal', 'Anomaly'], rotation=0)
    
    df['IsAnomaly'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%',
                                        colors=['#2ecc71', '#e74c3c'],
                                        labels=['Normal', 'Anomaly'])
    axes[1].set_title('Class Proportion', fontsize=14, fontweight='bold')
    axes[1].set_ylabel('')
    
    plt.tight_layout()
    plt.savefig(f'{OUTPUT_DIR}/class_distribution.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("\n✓ Saved: class_distribution.png")
```

```python
# Feature correlation
feature_cols = [col for col in EXPECTED_FEATURES if col in df.columns]

plt.figure(figsize=(14, 10))
correlation_matrix = df[feature_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            fmt='.2f', square=True, linewidths=1)
plt.title('Feature Correlation Heatmap', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Saved: correlation_heatmap.png")
```

```python
# PCA visualization
if 'IsAnomaly' in df.columns:
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(df[feature_cols])
    
    plt.figure(figsize=(12, 6))
    scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], 
                         c=df['IsAnomaly'], cmap='coolwarm', 
                         alpha=0.6, s=10, edgecolors='k', linewidth=0.1)
    plt.colorbar(scatter, label='IsAnomaly')
    plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)')
    plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)')
    plt.title('PCA Visualization (2D)', fontsize=14, fontweight='bold')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(f'{OUTPUT_DIR}/pca_visualization.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("✓ Saved: pca_visualization.png")
```

---
## 3. Data Preparation

Prepare train/test split for modeling.

```python
# Prepare features and target
feature_cols = [col for col in EXPECTED_FEATURES if col in df.columns]
X = df[feature_cols].values

if 'IsAnomaly' in df.columns:
    y = df['IsAnomaly'].values
    
    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y
    )
    
    print(f"Training set: {X_train.shape[0]} samples")
    print(f"Test set: {X_test.shape[0]} samples")
    print(f"\nTraining anomaly rate: {y_train.mean() * 100:.2f}%")
    print(f"Test anomaly rate: {y_test.mean() * 100:.2f}%")
else:
    X_train = X
    X_test = None
    y_train = None
    y_test = None
    print(f"Total samples: {X.shape[0]}")
    print("No labels - proceeding with unsupervised learning only")
```

---
## 4. Random Forest (Supervised Learning)

Train and evaluate Random Forest classifier.

```python
if y_train is not None:
    print("=" * 80)
    print("RANDOM FOREST CLASSIFIER")
    print("=" * 80)
    
    # Train model
    rf_model = RandomForestClassifier(
        n_estimators=RF_N_ESTIMATORS,
        max_depth=RF_MAX_DEPTH,
        min_samples_split=5,
        min_samples_leaf=2,
        random_state=RANDOM_STATE,
        n_jobs=-1,
        class_weight='balanced'
    )
    
    print("\nTraining Random Forest...")
    rf_model.fit(X_train, y_train)
    print("✓ Training complete")
    
    # Predictions
    y_train_pred = rf_model.predict(X_train)
    y_test_pred = rf_model.predict(X_test)
    y_test_proba = rf_model.predict_proba(X_test)[:, 1]
    
    # Metrics
    print("\n" + "=" * 60)
    print("PERFORMANCE METRICS")
    print("=" * 60)
    print("\nTraining Set:")
    print(f"  Accuracy:  {accuracy_score(y_train, y_train_pred):.4f}")
    print(f"  F1-Score:  {f1_score(y_train, y_train_pred):.4f}")
    
    print("\nTest Set:")
    print(f"  Accuracy:  {accuracy_score(y_test, y_test_pred):.4f}")
    print(f"  Precision: {precision_score(y_test, y_test_pred):.4f}")
    print(f"  Recall:    {recall_score(y_test, y_test_pred):.4f}")
    print(f"  F1-Score:  {f1_score(y_test, y_test_pred):.4f}")
    print(f"  ROC-AUC:   {roc_auc_score(y_test, y_test_proba):.4f}")
    
    print("\n" + "=" * 60)
    print("CLASSIFICATION REPORT")
    print("=" * 60)
    print(classification_report(y_test, y_test_pred, target_names=['Normal', 'Anomaly']))
else:
    print("⚠ No labels available for supervised learning")
```

```python
# Visualize Random Forest results
if y_train is not None:
    fig = plt.figure(figsize=(18, 10))
    
    # 1. Confusion Matrix
    ax1 = plt.subplot(2, 3, 1)
    cm = confusion_matrix(y_test, y_test_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax1,
                xticklabels=['Normal', 'Anomaly'], yticklabels=['Normal', 'Anomaly'])
    ax1.set_title('Confusion Matrix', fontsize=14, fontweight='bold')
    ax1.set_ylabel('True Label')
    ax1.set_xlabel('Predicted Label')
    
    # 2. ROC Curve
    ax2 = plt.subplot(2, 3, 2)
    fpr, tpr, _ = roc_curve(y_test, y_test_proba)
    auc_score = roc_auc_score(y_test, y_test_proba)
    ax2.plot(fpr, tpr, label=f'ROC (AUC = {auc_score:.3f})', linewidth=2.5)
    ax2.plot([0, 1], [0, 1], 'k--', label='Random', linewidth=1.5)
    ax2.set_xlabel('False Positive Rate')
    ax2.set_ylabel('True Positive Rate')
    ax2.set_title('ROC Curve', fontsize=14, fontweight='bold')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # 3. Precision-Recall Curve
    ax3 = plt.subplot(2, 3, 3)
    precision, recall, _ = precision_recall_curve(y_test, y_test_proba)
    ax3.plot(recall, precision, linewidth=2.5)
    ax3.set_xlabel('Recall')
    ax3.set_ylabel('Precision')
    ax3.set_title('Precision-Recall Curve', fontsize=14, fontweight='bold')
    ax3.grid(True, alpha=0.3)
    
    # 4. Feature Importance
    ax4 = plt.subplot(2, 3, 4)
    feature_importance = pd.DataFrame({
        'feature': feature_cols,
        'importance': rf_model.feature_importances_
    }).sort_values('importance', ascending=False)
    
    top_features = feature_importance.head(10)
    ax4.barh(range(len(top_features)), top_features['importance'], color='steelblue')
    ax4.set_yticks(range(len(top_features)))
    ax4.set_yticklabels(top_features['feature'])
    ax4.set_xlabel('Importance')
    ax4.set_title('Top 10 Feature Importances', fontsize=14, fontweight='bold')
    ax4.invert_yaxis()
    
    # 5. Prediction Distribution
    ax5 = plt.subplot(2, 3, 5)
    ax5.hist(y_test_proba[y_test == 0], bins=50, alpha=0.7, label='Normal', color='green', density=True)
    ax5.hist(y_test_proba[y_test == 1], bins=50, alpha=0.7, label='Anomaly', color='red', density=True)
    ax5.set_xlabel('Predicted Probability')
    ax5.set_ylabel('Density')
    ax5.set_title('Prediction Probability Distribution', fontsize=14, fontweight='bold')
    ax5.legend()
    ax5.grid(True, alpha=0.3)
    
    # 6. Metrics Summary
    ax6 = plt.subplot(2, 3, 6)
    metrics = {
        'Accuracy': accuracy_score(y_test, y_test_pred),
        'Precision': precision_score(y_test, y_test_pred),
        'Recall': recall_score(y_test, y_test_pred),
        'F1-Score': f1_score(y_test, y_test_pred),
        'ROC-AUC': auc_score
    }
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
    ax6.bar(metrics.keys(), metrics.values(), color=colors)
    ax6.set_ylim([0, 1])
    ax6.set_title('Performance Metrics', fontsize=14, fontweight='bold')
    ax6.set_ylabel('Score')
    ax6.tick_params(axis='x', rotation=45)
    for i, (k, v) in enumerate(metrics.items()):
        ax6.text(i, v + 0.02, f'{v:.3f}', ha='center', fontweight='bold')
    
    plt.tight_layout()
    plt.savefig(f'{OUTPUT_DIR}/random_forest_results.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("✓ Saved: random_forest_results.png")
```

---
## 5. Isolation Forest (Unsupervised Learning)

Train and evaluate Isolation Forest for anomaly detection.

```python
print("=" * 80)
print("ISOLATION FOREST")
print("=" * 80)

# Train model
iso_forest = IsolationForest(
    contamination=ISO_CONTAMINATION,
    random_state=RANDOM_STATE,
    n_estimators=100,
    max_samples='auto',
    n_jobs=-1
)

print(f"\nTraining Isolation Forest (contamination={ISO_CONTAMINATION})...")
iso_forest.fit(X_train)
print("✓ Training complete")

# Predictions
predictions = iso_forest.predict(X_train)
scores = iso_forest.score_samples(X_train)
anomalies = (predictions == -1).astype(int)

print(f"\nDetected anomalies: {anomalies.sum()} ({anomalies.mean() * 100:.2f}%)")

if y_train is not None:
    print("\n" + "=" * 60)
    print("PERFORMANCE ON LABELED DATA")
    print("=" * 60)
    print(f"Accuracy:  {accuracy_score(y_train, anomalies):.4f}")
    print(f"Precision: {precision_score(y_train, anomalies):.4f}")
    print(f"Recall:    {recall_score(y_train, anomalies):.4f}")
    print(f"F1-Score:  {f1_score(y_train, anomalies):.4f}")
```

```python
# Visualize Isolation Forest results
fig = plt.figure(figsize=(18, 10))

# 1. Anomaly Score Distribution
ax1 = plt.subplot(2, 3, 1)
ax1.hist(scores, bins=50, edgecolor='black', alpha=0.7, color='steelblue')
threshold = scores[predictions == -1].max()
ax1.axvline(threshold, color='red', linestyle='--', linewidth=2,
           label=f'Anomaly Threshold: {threshold:.4f}')
ax1.set_xlabel('Anomaly Score')
ax1.set_ylabel('Frequency')
ax1.set_title('Anomaly Score Distribution', fontsize=14, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. PCA with Anomalies
ax2 = plt.subplot(2, 3, 2)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train)

scatter = ax2.scatter(X_pca[:, 0], X_pca[:, 1], 
                     c=anomalies, cmap='coolwarm', 
                     alpha=0.6, s=10, edgecolors='k', linewidth=0.1)
ax2.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%})')
ax2.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%})')
ax2.set_title('PCA with Detected Anomalies', fontsize=14, fontweight='bold')
plt.colorbar(scatter, ax=ax2, label='Anomaly')

# 3. Score vs Sample (subsampled for visibility)
ax3 = plt.subplot(2, 3, 3)
max_points = min(10000, len(scores))
if len(scores) > max_points:
    plot_idx = np.random.choice(len(scores), max_points, replace=False)
    plot_idx.sort()
else:
    plot_idx = np.arange(len(scores))

ax3.scatter(plot_idx, scores[plot_idx], c=anomalies[plot_idx], cmap='coolwarm', 
           alpha=0.5, s=1)
ax3.set_xlabel('Sample Index')
ax3.set_ylabel('Anomaly Score')
ax3.set_title('Anomaly Scores Over Samples', fontsize=14, fontweight='bold')
ax3.grid(True, alpha=0.3)

# 4. Confusion Matrix (if labels available)
if y_train is not None:
    ax4 = plt.subplot(2, 3, 4)
    cm = confusion_matrix(y_train, anomalies)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges', ax=ax4,
                xticklabels=['Normal', 'Anomaly'], yticklabels=['Normal', 'Anomaly'])
    ax4.set_title('Confusion Matrix', fontsize=14, fontweight='bold')
    ax4.set_ylabel('True Label')
    ax4.set_xlabel('Predicted Label')

# 5. Top Anomalies
ax5 = plt.subplot(2, 3, 5)
top_n = min(100, len(scores))
top_anomaly_indices = np.argsort(scores)[:top_n]
ax5.scatter(range(top_n), scores[top_anomaly_indices], 
           color='red', alpha=0.7, s=30)
ax5.set_xlabel('Rank')
ax5.set_ylabel('Anomaly Score')
ax5.set_title(f'Top {top_n} Anomalies by Score', fontsize=14, fontweight='bold')
ax5.grid(True, alpha=0.3)

# 6. Summary
ax6 = plt.subplot(2, 3, 6)
summary_text = f"""
Total Samples: {len(predictions):,}

Detected Anomalies: {anomalies.sum():,}
Detection Rate: {anomalies.mean() * 100:.2f}%

Score Range: [{scores.min():.4f}, {scores.max():.4f}]
Mean Score: {scores.mean():.4f}
"""

if y_train is not None:
    acc = accuracy_score(y_train, anomalies)
    f1 = f1_score(y_train, anomalies)
    summary_text += f"\nAccuracy: {acc:.4f}\nF1-Score: {f1:.4f}"

ax6.text(0.5, 0.5, summary_text, ha='center', va='center', 
        fontsize=12, family='monospace',
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.5))
ax6.axis('off')
ax6.set_title('Detection Summary', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig(f'{OUTPUT_DIR}/isolation_forest_results.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Saved: isolation_forest_results.png")
```

---
## 6. Export Basic Predictions

Save predictions to CSV files.

```python
if y_train is not None:
    # Random Forest predictions
    rf_results = pd.DataFrame({
        'True_Label': y_test,
        'RF_Predicted': y_test_pred,
        'RF_Probability': y_test_proba,
        'RF_Correct': y_test == y_test_pred
    })
    rf_results.to_csv(f'{OUTPUT_DIR}/rf_predictions.csv', index=False)
    print("✓ Saved: rf_predictions.csv")

# Isolation Forest predictions
iso_results = pd.DataFrame({
    'IsoForest_Predicted': anomalies,
    'IsoForest_Score': scores
})

if y_train is not None:
    iso_results['True_Label'] = y_train
    iso_results['IsoForest_Correct'] = y_train == anomalies

iso_results.to_csv(f'{OUTPUT_DIR}/iso_predictions.csv', index=False)
print("✓ Saved: iso_predictions.csv")
```

---
## 7. Results Summary

Compare all models and display final metrics.

```python
if y_train is not None:
    print("=" * 80)
    print("MODEL COMPARISON SUMMARY")
    print("=" * 80)
    
    comparison = pd.DataFrame({
        'Model': ['Random Forest', 'Isolation Forest'],
        'Accuracy': [
            accuracy_score(y_test, y_test_pred),
            accuracy_score(y_train, anomalies)
        ],
        'Precision': [
            precision_score(y_test, y_test_pred),
            precision_score(y_train, anomalies)
        ],
        'Recall': [
            recall_score(y_test, y_test_pred),
            recall_score(y_train, anomalies)
        ],
        'F1-Score': [
            f1_score(y_test, y_test_pred),
            f1_score(y_train, anomalies)
        ]
    })
    
    print("\n", comparison.to_string(index=False))
    
    # Visualize comparison
    fig, ax = plt.subplots(figsize=(12, 6))
    
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
    x = np.arange(len(metrics))
    width = 0.35
    
    rf_scores = comparison[comparison['Model'] == 'Random Forest'][metrics].values[0]
    iso_scores = comparison[comparison['Model'] == 'Isolation Forest'][metrics].values[0]
    
    ax.bar(x - width/2, rf_scores, width, label='Random Forest', color='steelblue')
    ax.bar(x + width/2, iso_scores, width, label='Isolation Forest', color='coral')
    
    ax.set_ylabel('Score', fontsize=12)
    ax.set_title('Model Comparison', fontsize=14, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(metrics)
    ax.legend()
    ax.set_ylim([0, 1])
    ax.grid(True, alpha=0.3, axis='y')
    
    for i, (rf, iso) in enumerate(zip(rf_scores, iso_scores)):
        ax.text(i - width/2, rf + 0.02, f'{rf:.3f}', ha='center', fontsize=9)
        ax.text(i + width/2, iso + 0.02, f'{iso:.3f}', ha='center', fontsize=9)
    
    plt.tight_layout()
    plt.savefig(f'{OUTPUT_DIR}/model_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("\n✓ Saved: model_comparison.png")
```
